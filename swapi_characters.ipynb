```python
import requests
import pandas as pd

# Fetch page 1
url = "https://swapi.dev/api/people/"
response = requests.get(url).json()

characters = response["results"]
next_url = response["next"]

# Fetch page 2 to get 20 characters total
response2 = requests.get(next_url).json()
characters += response2["results"]

df = pd.DataFrame([
    {
        "name": c["name"],
        "height": c["height"],
        "mass": c["mass"],
        "gender": c["gender"],
        "birth_year": c["birth_year"],
        "homeworld_url": c["homeworld"],
        "films_count": len(c["films"])
    }
    for c in characters
])

df
```

```python
df.to_csv("swapi_characters.csv", index=False)
df
```
